# User Report

```python
import pandas as pd
import intralinks
import intralinks.tools.login_manager

il = intralinks.IntralinksClient()

login_manager = intralinks.tools.login_manager.LoginManager(il)

ui = intralinks.tools.login_manager.LoginManagerUI(login_manager)

ui.login_panel()
```

```python
exchanges = il.get_exchanges()

exchanges = [e for e in exchanges if e['type'] != 'IL5' and e['id'] != 1353755]
```

# Users only

```python
all_users = []

for exchange in exchanges:
    print('Retrieving users for exchange #{}'.format(exchange['id']))
    users = il.get_exchange_members(exchange['id'])

    for user in users:
        all_users.append({'user':user, 'exchange':exchange})
```

```python
import pandas
import pandas.io.json

dataframe = pandas.DataFrame(pandas.io.json.json_normalize(all_users))

dataframe[[
    'user.firstName', 
    'user.lastName', 
    'user.organization', 
    'user.emailId', 
    'user.roleType', 
    'user.lastAccessedOn.milliseconds', 
    'user.lastAlertSentDate.milliseconds', 
    'user.createdBy.firstName', 
    'user.createdBy.lastName', 
    'user.createdOn.milliseconds', 
    'exchange.id', 
    'exchange.workspaceName'
]]
```

# Users and groups

```python
import intralinks.utils.associations

all_users = []

for exchange in exchanges:
    print('Retrieving users & groups for exchange #{}'.format(exchange['id']))
    
    groups, group_members = il.get_groups_and_members(exchange['id'])
    users = il.get_exchange_members(exchange['id'])
    intralinks.utils.associations.associate_users_and_groups(users, groups, group_members)

    groups_by_id = { g['id']:g for g in groups }

    for user in users:            
        if len(user['groups']) > 0:
            for group_id in user['groups']:
                group = groups_by_id[group_id]
                all_users.append({'user':user, 'group':group, 'exchange':exchange})
        else:
            all_users.append({'user':user, 'group':None, 'exchange':exchange})
```

```python
import pandas

dataframe = pandas.DataFrame(pandas.io.json.json_normalize(all_users))

dataframe[[
    'user.firstName', 
    'user.lastName', 
    'user.organization', 
    'user.emailId', 
    'user.roleType', 
    'user.lastAccessedOn.milliseconds', 
    'user.lastAlertSentDate.milliseconds', 
    'user.createdBy.firstName', 
    'user.createdBy.lastName', 
    'user.createdOn.milliseconds',
    'group.id',
    'group.groupName',
    'group.groupType',
    'exchange.id', 
    'exchange.workspaceName'    
]]
```
